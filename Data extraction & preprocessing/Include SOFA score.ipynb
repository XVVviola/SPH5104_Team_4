```python
import pandas as pd
import numpy as np
```

```python
# Load patietn data
# n=1631 patients 

df = pd.read_csv('Final data_9April.csv')
df
```

```python
# Load SOFA score dataset for 1631 patients
# SOFA score data extracted based on stay_id

sofa = pd.read_csv('Final study cohort_SOFA scores.csv') 
sofa.head(5)
```

```python
# sort by hr
sofa.sort_values(by='hr',ascending=True,inplace=True)
```

#### Include Base SOFA score - first SOFA score

```python
# Include only the first sofa score for each patient - baseline sofa score 

Base_sofa = sofa.groupby('stay_id').first().reset_index()
Base_sofa
```

```python
# Include baseline sofa score into the final study cohort

df_final = pd.merge(df,Base_sofa[['stay_id','sofa_24hours']],how='left',on=['stay_id'])
```

```python
df_final.to_csv('Final cohort with SOFA.csv',index=False)
```
